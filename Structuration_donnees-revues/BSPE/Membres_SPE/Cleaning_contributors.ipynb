```python
import csv
import re
import numpy as np
```

```python
f = open ('R_contributeurs-1908-1921.csv', 'r', encoding='utf16')
```

```python
csv_f = csv.reader(f)
```

```python
contributeurs_listes = []
for row in csv_f :
    contributeurs_listes.append(row)
print(contributeurs_listes)
#Problème : chaque nom est une liste à lui même
```

```python
##Faire une seule liste constituée de tous les noms
flat_list = []

for sublist in contributeurs_listes:
    for item in sublist:
        flat_list.append(item)
#print(flat_list)
len(flat_list)
```

```python
contributeurs_clean = []

for i in range (len(flat_list)-1) :
    if flat_list[i] != flat_list[i+1] :
        contributeurs_clean.append(flat_list[i])
print(contributeurs_clean)
```

```python
print ("nombre de contributeurs : ",len(contributeurs_clean))
print("\n")
print("Exemple avec le nombre d'occurences de Yakimoff :")
print("avant nettoyage : ",flat_list.count('YAKIMOFF (W.L.)'))
print("après nettoyage : ",contributeurs_clean.count('YAKIMOFF (W.L.)'))
```
